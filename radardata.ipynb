```python
import pandas as pd
from statistics import mean
import numpy as np

df = pd.read_csv('data.csv', sep = ',')

df
```

```python
    
col = ['whose',
      'danceability',
      'energy',
      'acousticness',
      'instrumantalness',
      'liveness',
      'valence'
      ]
```

```python
unique = np.unique(df['whose'])
spider = pd.DataFrame(index = [i for i in range(3)], columns = col)

j = 0
for i in unique:
    row = np.array([
        i,
        mean(df[df['whose'] == i]['danceability']),
        mean(df[df['whose'] == i]['energy']),
        mean(df[df['whose'] == i]['acousticness']),
        mean(df[df['whose'] == i]['instrumentalness']),
        mean(df[df['whose'] == i]['liveness']),
        mean(df[df['whose'] == i]['valence'])
    ])
    
    spider[j:(j+1)] = row.reshape(1, 7)
    j += 1
    
    
spider
```
